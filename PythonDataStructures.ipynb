```python
#Sort the frequency counter using the sorting algorithms provided in the table below.
#Additionally, for each of the sorting methods, leverage the time function in Python to see how many milliseconds it takes for an algorithm to sort items.
#Rank the sorting algorithms in order of the time they take to sort items.


frequency_dict = {}  # creating the empty dictionary

frequency_list =[] # since dictionary cannot be sorted, we need to convert the dictionary into a list. Hence assigning an empty list

mergesort_frequency_dict = {} # to hold the sorted dictionary ordered by merge sort

insertionsort_frequency_dict = {} # to hold the sorted dictionary ordered by insertion sort

selectionsort_frequency_dict = {} # to hold the sorted dictionary ordered by selection sort

time_calc_dict = {} # to capture the time informtion for each sort method

for transaction in transaction_data: # looping over each transaction
    for item in transaction: # looping over each item present in the transaction
        if item in frequency_dict:
            frequency_dict[item] += 1 # increment the frequency by 1 if the item is found in the dictionary
        else:
            frequency_dict[item] = 1 # create a new key with the item name if the item is found for the first time

frequency_list = list(frequency_dict.items()) # extracting the vlaues from frequency counter as a list

# Merge Sort

# Defining the function for merge sort
def mergeSort(frequency_list):
    
    
    # if the length of the array is less than 2, then we do not need to sort
    if len(frequency_list) < 2:
        return frequency_list
    
    result = [] # assigning the empty list to store the result
    
    mid = int(len(frequency_list) / 2) # we need to determine and mid point of the array to divide  in two halves
    
    l = mergeSort(frequency_list[:mid]) # left side of the mid point 
    
    r = mergeSort(frequency_list[mid:]) # right side of the mid point
    
    # two variables for the counter
    i = 0
    j = 0
    
    # the while loop runs till the leghth of the both left and right array
    while i < len(l) and j < len(r):
        
        if l[i][1] < r[j][1]: # chacking for each value from left array whether this is less than value from right array 
            result.append(r[j]) # if left is less than right, add the value from right array in the result. this is beacuse we want to order in descending order
            j += 1 
        else:
            result.append(l[i]) # if not we add the value from left array 
            i += 1
    result += l[i:] # add the remaining values from left array, if any
    result += r[j:] # add the remaining values from right array, if any
    

    # returning result       
    return result

# assigning empty variables to store the time calculation
start_time = 0
time_taken = 0
sort_method = 'Merge Sort'


# calculate the start time to measure the perfomance of the sort
start_time = time.time()*1000

# driver code to call the merge sort function 
mergesort_frequency_dict = dict(mergeSort(frequency_list))

# calculating the time taken 
time_taken = (time.time()*1000 - start_time)

# adding the result to a dictionary
time_calc_dict[sort_method] = time_taken



print("Using Merge Sort Method")
print("=======================")
print("Item ", " Frequency")# printing the header of the frequency counter


for k, v in mergesort_frequency_dict.items(): # looping through the sorted dictionary
    print(k.capitalize(), v)
    
    

    
# Insertion sort

# Defining the function for insertion sort
def insertionSort(frequency_list):

    for i in range(1, len(frequency_list)):
        key = frequency_list[i] # assigning the 'i'th index for key 
        j = i - 1
        
        # Comparing key with each element on the left of it until an element greater than it is found       
        while j >= 0 and key[1] > frequency_list[j][1]:
            frequency_list[j + 1] = frequency_list[j]
            j = j - 1
        
        # Place key at after the element
        frequency_list[j + 1] = key
    
    # returning result 
    return frequency_list

# assigning empty variables to store the time calculation
start_time = 0
time_taken = 0
sort_method = 'Insertion Sort'


# calculate the start time to measure the perfomance of the sort
start_time = time.time()*1000

# driver code to call the insertion sort function 
insertionsort_frequency_dict = dict(insertionSort(frequency_list))

# calculating the time taken 
time_taken = (time.time()*1000 - start_time)

# adding the result to a dictionary
time_calc_dict[sort_method] = time_taken


print('                           ')
print("Using Insertion Sort Method")
print("===========================")
print("Item ", " Frequency")# printing the header of the frequency counter


for k, v in insertionsort_frequency_dict.items(): # looping through the sorted dictionary
    print(k.capitalize(), v)


    

# Selection sort

# Defining the function for selection sort
def selectionSort(frequency_list):
    
    for i in range(len(frequency_list)): # for the entire length of the frequency list
        max_idx = i # assigning th 'i'th elemnet as the maximum index

        for j in range(i + 1, len(frequency_list)):
         
            # select the bigger element in each loop
            if frequency_list[j][1] > frequency_list[max_idx][1]:
                max_idx = j
         
        # put greater element at the right position
        (frequency_list[i], frequency_list[max_idx]) = (frequency_list[max_idx], frequency_list[i])
    
    # returning result 
    return frequency_list

# assigning empty variables to store the time calculation
start_time = 0
time_taken = 0
sort_method = 'Selection Sort'


# calculate the start time to measure the perfomance of the sort
start_time = time.time()*1000

# driver code to call the insertion sort function 
selectionsort_frequency_dict = dict(selectionSort(frequency_list))

# calculating the time taken 
time_taken = (time.time()*1000 - start_time)

# adding the result to a dictionary
time_calc_dict[sort_method] = time_taken


print('                           ')
print("Using Selection Sort Method")
print("===========================")
print("Item ", " Frequency")# printing the header of the frequency counter


for k, v in selectionsort_frequency_dict.items(): # looping through the sorted dictionary
    print(k.capitalize(), v)
    

    

    
# Printing the time taken

print('                           ')
print("Time Calculation for Algorithms")
print("===========================")
print("Algorithm ", " Time (in milliseconds)")# printing the header of the frequency counter

time_calc_dict = dict(sorted(time_calc_dict.items(), key=lambda item: item[1]))

for k, v in time_calc_dict.items(): # looping through the sorted dictionary
    print(k, v)
```
